```sql
select u.[LOGIN],
       u.NEV,
       f.METTOL,
       f.MEDDIG,
       f.FELNOTT_SZAM+f.GYERMEK_SZAM as 'személyek száma',
       sum(f.FELNOTT_SZAM+f.GYERMEK_SZAM) 
        over (partition by u.[LOGIN] order by u.[LOGIN], f.METTOL rows between 3 preceding and 1 preceding)
from Foglalas f join Ugyfel u on f.UGYFEL_FK = u.[LOGIN]
```

```sql
SELECT rt.TERMEKKOD, 
       AVG(rt.EGYSEGAR) AS 'Átlagos ár'
FROM Rendeles_tetel rt 
CROSS APPLY 
(select * from Rendeles r
     where rt.SORSZAM = r.SORSZAM)r
GROUP BY rt.TERMEKKOD
```
